## Relationship & Correlation Analysis 

 - Is BMI correlated with insurance charges? 
 - Is age correlated with insurance charges? 
 - Does number of children correlate with charges?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data\insurance.csv")
df.head()
```

1) Is BMI correlated with insurance charges?

```python
df['bmi'].corr(df['charges'])
```

```python
plt.scatter(df['bmi'], df['charges'])
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("BMI vs Charges")
plt.show()
```

2) Is age correlated with insurance charges?

```python
df['age'].corr(df['charges'])
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['age'], df['charges'])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Charges")
plt.show()
```

3.Does number of children correlate with charges?

```python
df['children'].corr(df['charges'])
```

```python
plt.scatter(df['children'], df['charges'])
plt.xlabel("Children")
plt.ylabel("Charges")
plt.title("Children vs Charges")
plt.show()
```

Correlation analysis shows that age and BMI both have positive relationships with insurance charges, with age showing a slightly stronger effect than BMI. However, the number of children has very weak correlation with charges, indicating it is not a significant factor in determining insurance cost.
