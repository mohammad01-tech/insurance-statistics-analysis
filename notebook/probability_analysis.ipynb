# Probability Analysis 
- Understanding likelihood of events. 
- What is the probability that a customer has high insurance charges?
- What is the probability that a customer is a smoker? 
- What is the probability that a high-cost customer is a smoker? 
- What is the probability that a high-cost customer is obese (high BMI)?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data\insurance.csv")
df.head()
```

1) Probability that a customer has high insurance charges

```python
threshold = df['charges'].quantile(0.9)
high_cost = df[df['charges'] >= threshold]
p_high_cost_charges = len(high_cost)/len(df)*100
p_high_cost_charges
```

2) Probability that a customer is a smoker

```python
(df['smoker'].value_counts(normalize=True)*100)
```


3) Probability that a high-cost customer is a smoker

```python
p_high_cost = (high_cost['smoker'].value_counts(normalize=True)*100)
p_high_cost
```

4) Probability that a high-cost customer is obese (BMI â‰¥ 30)

```python
high_cost_obese = high_cost[high_cost['bmi'] >=30]
p_cost_obese = len(high_cost_obese)/len(high_cost)*100
p_cost_obese
```

## REPORT
Probability analysis shows that only about 10% of customers fall into the high-cost category. However, among these high-cost customers, nearly all are smokers (97.76%) and obese (96.27%). This indicates that smoking and high BMI are the most critical risk factors driving insurance charges.
